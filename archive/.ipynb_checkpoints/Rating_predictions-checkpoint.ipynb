```python
# Data preprocessing, math and plotting
import os
import numpy as np 
import pandas as pd 
from scipy import stats 
import matplotlib.pyplot as plt
import seaborn as sns 
import datetime
import time
import random
import warnings
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
nltk.download('stopwords')
import warnings
warnings.filterwarnings('ignore')

# ML
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import sklearn
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.neighbors import NearestNeighbors, KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, auc, classification_report, r2_score, accuracy_score, confusion_matrix, roc_auc_score, roc_curve, precision_score, recall_score, f1_score  
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn import tree
from sklearn.utils import resample
```

```python
# HYPERPARAMETERS
RANDOM_STATE = 12
TRAIN_SPLIT = 0.80
OOV_TOK = "<OOV>"
VOCAB_SIZE = 10000
MAX_LENGTH = 120
MAX_ITER = 10000
N_JOBS = 4
N_NEIGHBORS = 5
N_ESTIMATORS = 50
MAX_DEPTH = 1
```

```python
data = pd.read_csv(r'olist_order_reviews_dataset.csv')
data.head()
```

```python
# Define a function called "clean_data" that takes several arguments:
# "data" is the dataframe to be cleaned
# "column_text" is the name of the column containing the text to be used for classification (default is "review_comment_message")
# "column_score" is the name of the column containing the rating score (default is "review_score")
# "points_cut" is a list of cutoff points for the rating score (default is [0, 2, 5])
# "classes" is a list of class labels corresponding to the cutoff points (default is [0, 1])

def clean_data(data, column_text='review_comment_message', 
               column_score='review_score', 
               points_cut = [0, 2, 5], 
               classes = [0, 1]):
    
    # Create a new dataframe called "df_bin" that is a copy of the input "data" dataframe
    df_bin = data
    
    # Remove rows with missing data in the "column_text" column
    df_bin = df_bin.dropna(subset=[column_text])
    
    # Create a new column called "label" based on the "column_score" column and the cutoff points
    # The "labels" parameter is used to assign class labels to each range of scores
    df_bin['label'] = pd.cut(df_bin[column_score], bins=points_cut, labels=classes)
    
    # Rename the "column_text" column to "text" and select only the "text" and "label" columns
    df_bin = df_bin.rename(columns={column_text: 'text'})
    df_bin = df_bin[['text','label']]
    
    # Create a new dataframe called "df_cat" that is a copy of the input "data" dataframe
    df_cat = data
    
    # Remove rows with missing data in the "column_text" column
    df_cat = df_cat.dropna(subset=[column_text])
    
    # Rename the "column_text" column to "text" and the "column_score" column to "label"
    df_cat = df_cat.rename(columns={column_text: 'text' , column_score: 'label'})
    
    # Select only the "text" and "label" columns
    df_cat = df_cat[['text','label']]
    
    # Return two dataframes, one for binary classification and one for categorical classification
    return df_bin ,df_cat

# Call the "clean_data" function with the "data" dataframe and assign the returned dataframes to "data_bin" and "data_cat"
data_bin , data_cat = clean_data(data)
```

# Machine Learning Pipeline

```python
def split_test_train(data, split_train=TRAIN_SPLIT, random_state=RANDOM_STATE):
    # Separate majority and minority classes
    df_majority = data[data['label']==0]
    df_minority = data[data['label']==1]

    # Undersample majority class
    n_samples = min(len(df_majority), len(df_minority))
    df_majority_downsampled = resample(df_majority, 
                                       replace=False,    # sample without replacement
                                       n_samples=n_samples, # match minority class
                                       random_state=random_state)

    # Combine minority class with downsampled majority class
    df_balanced = pd.concat([df_majority_downsampled, df_minority])

    # Split into train and test sets
    df_train = df_balanced.sample(frac=split_train, random_state=random_state)
    df_test = df_balanced.drop(df_train.index)
    X_train = df_train['text'].tolist()
    y_train = df_train['label'].values
    X_test = df_test['text'].tolist()
    y_test = df_test['label'].values

    return X_train , y_train , X_test , y_test

X_train , y_train , X_test , y_test = split_test_train(data_bin, split_train=TRAIN_SPLIT, random_state=RANDOM_STATE)
```

```python
# Import the Tokenizer class from the keras.preprocessing.text module and instantiate an object with specified parameters.
tokenizer = Tokenizer(num_words = VOCAB_SIZE, oov_token=OOV_TOK)

# Fit the tokenizer on the training set, creating a word index based on the frequency of occurrence of each word in the corpus.
tokenizer.fit_on_texts(X_train)

# Define a function to preprocess the text data by converting each sentence into a sequence of integers using the fitted tokenizer and padding/truncating them to a fixed length.
def preprocess(X_train, X_test, max_length, vocab_size, trunc_type='post', oov_tok = "<OOV>"):

    # Convert each sentence in the training set to a sequence of integers using the fitted tokenizer.
    training_sequences = tokenizer.texts_to_sequences(X_train)

    # Pad the sequences to a fixed length by adding zeros or truncating from the beginning or end based on the trunc_type parameter.
    X_train_padded = pad_sequences(training_sequences,maxlen=max_length, truncating=trunc_type)

    # Convert each sentence in the test set to a sequence of integers using the fitted tokenizer.
    testing_sequences = tokenizer.texts_to_sequences(X_test)

    # Pad the sequences to a fixed length without truncation since the test set should not be modified.
    X_test_padded = pad_sequences(testing_sequences,maxlen=max_length)

    # Return the padded training and test sets as output of the function.
    return X_train_padded, X_test_padded

# Call the preprocess function on the training and test sets to obtain padded and processed data.
X_train, X_test = preprocess(X_train, X_test, MAX_LENGTH, VOCAB_SIZE)
```

# Machine Learning Models and Model Performance Metrics

```python
models = [
    ('MultinomialNB', MultinomialNB()),
    ('LogisticRegression', LogisticRegression(max_iter=MAX_ITER, random_state=RANDOM_STATE, n_jobs=N_JOBS)),
    ('SVM', SVC(random_state=RANDOM_STATE, verbose=False, probability=True)),
    ('RandomForest', RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE, n_jobs=N_JOBS)),
    ('XGBoost', XGBClassifier(objective='binary:logistic', use_label_encoder=False, random_state=RANDOM_STATE, eval_metric='mlogloss')),
    ('AdaBoost', AdaBoostClassifier(random_state=RANDOM_STATE)),
    ('KNeighbors', KNeighborsClassifier(n_neighbors=N_NEIGHBORS)),
    ('MLP', MLPClassifier(random_state=RANDOM_STATE, verbose=False)),
    ('SGD', SGDClassifier(random_state=RANDOM_STATE, verbose=False))
]

for name, model in models:
    start_time = time.time()
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    score_train = model.score(X_train, y_train) * 100
    score_test = model.score(X_test, y_test) * 100
    end_time = time.time()
    print(f"{name} - Train Score: {score_train:.2f}%, Test Score: {score_test:.2f}%, Time: {end_time - start_time:.2f}s")

metrics = {}

for name, model in models:
    if name == "SGD":
        pred = model.decision_function(X_test)
    else:
        pred = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, pred)
    roc_auc = roc_auc_score(y_test, pred)
    accuracy = accuracy_score(y_test, model.predict(X_test))
    precision = precision_score(y_test, model.predict(X_test))
    recall = recall_score(y_test, model.predict(X_test))
    f1 = f1_score(y_test, model.predict(X_test))
    tn, fp, fn, tp = confusion_matrix(y_test, model.predict(X_test)).ravel()
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    metrics[name] = {"Accuracy": accuracy, "Precision": precision, "Recall": recall, "F1-Score": f1,
                     "Sensitivity": sensitivity, "Specificity": specificity, "FPR": fpr, "TPR": tpr, "ROC/AUC": roc_auc}

# Bar chart of performance metrics
metric_names = ["Accuracy", "Precision", "F1-Score", "Sensitivity", "Specificity", "ROC/AUC"]
num_metrics = len(metric_names)
num_models = len(models)

fig, ax = plt.subplots(figsize=(16, 10))
width = 0.8 / num_metrics

for i, metric in enumerate(metric_names):
    offset = (i - num_metrics / 2) * width
    values = [metrics[name][metric] for name, _ in models]
    ax.bar(np.arange(num_models) + offset, values, width=width, label=metric, align='center')

ax.set_xticks(np.arange(num_models) + width/2)  # Add half of the bar width
ax.set_xticklabels([name for name, _ in models], rotation=45, ha='right')
ax.tick_params(axis='x', which='major', pad=15)
ax.legend()
ax.set_title("Comparison of Model Performance Metrics", fontsize=22)

plt.show()
```

# Confusion Matrices

```python
# set the size of the figure
plt.figure(figsize=(20, 15))

# set the title of the figure
plt.suptitle("Confusion Matrices", fontsize=24)

# loop through each model and plot the confusion matrix
for i, (name, model) in enumerate(models):
    # generate the confusion matrix
    cm = confusion_matrix(y_test, model.predict(X_test))
    # create a subplot for each model
    plt.subplot(3, 3, i+1)
    # set the title of the subplot
    plt.title(name)
    # plot the heatmap for the confusion matrix
    sns.heatmap(cm, cbar=False, cmap="YlGn", annot=True, annot_kws={"size": 16}, fmt='g')

# adjust the spacing between the subplots
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
```

# Overlaid ROC curve for each model

```python
# Create a new figure with a specific size
plt.figure(figsize=(12, 10))

# Iterate over each model and plot its ROC curve
for name, model in models:
    # For the SGD model, use the decision function to make predictions
    if name == "SGD":
        pred = model.decision_function(X_test)
    # For other models, use predict_proba to get the probability of the positive class
    else:
        pred = model.predict_proba(X_test)[:, 1]
    # Compute the false positive rate, true positive rate, and thresholds for the ROC curve
    fpr, tpr, _ = roc_curve(y_test, pred)
    # Compute the AUC score for the ROC curve
    roc_auc = roc_auc_score(y_test, pred)
    # Plot the ROC curve for the model, with a label that includes the model name and AUC score
    plt.plot(fpr, tpr, label=f"{name} (AUC = {roc_auc:.2f})")

# Plot a diagonal line to represent the ROC curve of a random classifier
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
# Add labels for the x-axis, y-axis, and title of the plot
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves', fontsize=20)
# Add a legend to the plot in the lower right corner
plt.legend(loc='lower right')
# Display the plot
plt.show()
```
